### Without %matplotlib this we can't show plottings in note

```python
%matplotlib inline 
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,5,7])
plt.title("Dummy Data")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

## Looks Lit AF!

#### Plotting multiple data in a single chart

```python
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
```

```python
plt.plot(x,y, label="first")
plt.plot(x2,y2, label="second")
plt.legend()
plt.show()
```

#### That box in the upper right corner is called legend 

Now you can also change the style of graph using style.use("name")

```python
from matplotlib import style
style.use('ggplot')

x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y, label="first")
plt.plot(x2,y2, label="second")
plt.legend()
plt.show()
```
